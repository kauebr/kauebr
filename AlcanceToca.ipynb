# Análise de alcance das publições da página Toca de Pandora no facebook, durante os dois anos de existência da empresa, e do impacto do algorítmo de divulgação confidencial que criei.

*   Conjunto de dados retirado do Meta analytics da página do petshop e-commerce [Toca de Pandora](https://www.facebook.com/tocadepandora)
<hr>

# Criando o DataFrame

```python
import pandas as pd
df = pd.read_csv('AlcanceFbook.csv', encoding='latin1', sep=';')
pd.set_option('display.max_rows', 100)
pd.set_option('display.precision', 2)
display(df)
```

```python
total = df['Alcance'].sum()
reacoes = df['Curtidas e reações'].sum()
clicks = df['Cliques no link'].sum()
comen = df['Comentários'].sum()
comp = df['Compartilhamentos'].sum()

def top():
    print(33*'-')

top()
print(f"{'RESULTADO:':^33}")
top()
print(f'''Alcance total: {total}
Total de curtidas e reações: {reacoes}
Total de clicks no link: {clicks:.0f}
Total de comentários: {comen:.0f}
Total de compartilhamentos: {comp}''')
top()
```

# Conclusão

*   Os compartilhamentos foram direcionados a um público alvo específico: moradores de Curitiba e RMC: 3.223.863 habitantes.
* O alcance total foi de mais de 700 mil visualizações da nossa publicidade, porém este valor conta com intersecções, não são 700 mil pessoas, mas sim 700 mil vezes que a publicidade foi visualizada, podendo ter sido visualizada mais de uma vez pela mesma pessoa, todavia, obtivemos um bom alcance, como podemos observar no indicador de `clicks no link`, houveram `41.116 pessoas`(entradas únicas) que clicaram nos links de contato, O que representa `1,28%` da população total do nosso público alvo(Curitiba e RMC) entrando em contanto com nosso canal de atendimento. Isso gerou impactos comerciais positivos.
